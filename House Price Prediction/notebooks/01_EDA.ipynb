# 📊 Exploratory Data Analysis (EDA)
In Machine Learning, **Exploratory Data Analysis** is a critical first step. This notebook acts as your visual playground. Before writing clean, automated Python scripts in our `src/` folder, this is where we load the raw datasets manually, look at correlations, pinpoint missing values, and decide what features to use.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Prettify warnings and plots
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
```

## 1. Load the Data
We load the generic raw data (e.g. from Kaggle's House Price dataset). Make sure `train.csv` is inside `../data/raw/`.

```python
# Notice how we step BACK one directory "../" to reach the root data folder.
file_path = '../data/raw/train.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print(f"Dataset loaded successfully! Shape: {df.shape}")
else:
    print(f"⚠️ Data not found at {file_path}. Please drag and drop a dataset there!")
    # Providing a dummy frame so the notebook doesn't completely crash if no file exists
    df = pd.DataFrame({'SalePrice': [200000, 300000], 'GrLivArea': [1500, 2500]})
```

## 2. Basic Dataset Information
Look at what the data actually looks like, what the data types are, and if we have any massive holes (missing variables).

```python
display(df.head())
print("\n--- DataFrame Info ---")
display(df.info())
print("\n--- Missing Values (Top 10) ---")
display(df.isnull().sum().sort_values(ascending=False).head(10))
```

## 3. Target Variable Analysis
Let's look at the distribution of our Target (e.g., `SalePrice`). Usually, real-world metrics like house prices are heavily right-skewed.

```python
target_col = 'SalePrice'
if target_col not in df.columns:
    target_col = df.columns[-1] # Fallback to the last column if SalePrice isn't found
    
plt.figure(figsize=(10, 6))
sns.histplot(df[target_col], kde=True, color='purple', bins=40)
plt.title(f'Distribution of {target_col}')
plt.xlabel(f'{target_col} ($)')
plt.ylabel('Frequency (Number of Occurrences)')
plt.show()
```

## 4. Feature Correlation Matrix
Calculates mathematical correlations between every variable in the dataset. `1.0` is a perfect match. Which numeric features highly impact the target price?

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 10))

# Calculate Correlation matrix
corr = numeric_df.corr()

# Display Top 10 correlated features on a Heatmap
top_correlations = corr.nlargest(10, target_col)[target_col].index
cm = np.corrcoef(numeric_df[top_correlations].values.T)

sns.heatmap(cm, annot=True, square=True, fmt='.2f', 
            yticklabels=top_correlations.values, xticklabels=top_correlations.values,
            cmap='coolwarm', linewidths=0.5)
plt.title(f'Top Features Most Highly Correlated with {target_col}', fontsize=16)
plt.tight_layout()
plt.show()
```
